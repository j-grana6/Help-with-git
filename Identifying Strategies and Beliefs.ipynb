### So how do we take this into account for practical purposes?

Let's look at an example (from Information and Entropy Econometrics \- a Review and Synthesis. (2008) Golan)

Suppose you have a urn with blue, yellow and red balls.  Suppose 

* $P(Blue)=\frac{1}{2}$   
* $P(Yellow) = \frac{1}{4}$ 
* $P(Red)=\frac{1}{4}$

Someone else draws a ball and you want to determine its color just by asking yes or no questions.  What do you do?

$\mathbf{E}(\\# Questions |Blue, Red) = \frac{1}{2}\times 1 + \frac{1}{2} \times 2=\frac{3}{2}$

<br \>

$\mathbf{E}(\\# Questions |Red, Blue) = \frac{1}{4}\times 1 + \frac{3}{4} \times 2=\frac{7}{4}>\frac{3}{2}$

<br \> 
Note that the higher the minimum expected number of questions we have to ask, the less we know about a random variable.  Examples include:

* Winner of the 2013 World Series
* Percent of underwater mortgages
* The first person in the class to ask a question that I have no idea how to answer (although by now this RV may have already been determined).

### So what does this "minimum expected # of questions" thing have to do with information and entropy?

Shannon's Entropy:
$\mathbf{H}(p) = \displaystyle\sum_i p_i \log_2(p_i)$

Interestingly, the minimum expected number of questions to determine the value of a random variable, $X$ is between $\mathbf{H}(X) \;\;$   and   $\;\; \mathbf{H}(X) +1 \; \;$.  Therefore, the higher the entropy, the *more* uncertain we are.

### So we want to be as certain as possible so we will minimize the entropy, right?

Wrong!!!   Remember Bernoulli and the principle of insufficient reason.  We do not want to be more certain than our information warrants.  Imagine if a gambler *minimized* entropy of choosing the team that will win the world series.   

## Finally...the paper

### The Game

* A finite set of players $I = \\{1,...,I\\}$
* For each player $i$ a set of $\kappa_i$ pure strategies
* For each player $i$, a set of $\tau_i$ possible types $\Theta_i$
* A set of possible games $\Theta = \Theta_1 \times...\times \Theta_I$
* For each player, a type-contigent payoff function $u_i(\cdot| \theta_i): X \rightarrow \mathbf{R}$

### What's missing? What's known? 

* Everything!
* Nothing except
    - Strategies *might* depend on utility
    - A single instance of play

### What are we interested in?

* Type contingent strategies.  "Given player $i$ is type $\theta_i$, what is the probability she plays $\kappa_i$"

* Player $i$'s (possibly unique) prior distribution on the game being played $\alpha_i$.  

### So what is imposed?

$\displaystyle\sum_{\theta} \displaystyle\sum_x P_{\theta_i}(x_i)P_{\theta_j}(x_j)\alpha_i(\theta)u_i(x|\theta_i) = \displaystyle\sum_{\theta} \displaystyle\sum_x Y_i(x_i)Y_j(x_j)\alpha_i(\theta)u_i(x|\theta_i)$

Interpretation:  The 'true' expected utility of playing against *each of the players individually* is equal to the observed expcted utility of playing against *each of the players individually*.  **NO BEHAVIORAL ASSUMPTIONS!**

### Identification

Note that by definition, the game is not under\-identified.  However, it is much more likely to be under\-identified.

* $I \times (I-1) \; \; $ pairwise constraints $\;\; + I\;\;$ constraints on the prior distributions $+ \tau\;\;$ constraints to ensure that the type contingent strategies are proper distributions.  $ \;\;I \times (I-1) +I +\tau = I \times I +\tau$.
* $\displaystyle\sum_i \tau_i \times \kappa_i \;\;$ strategy probabilities $\; + I \times |\Theta| \;\;$ belief paramteres. 


### The Maximization Problem

#### Entropy:

$H(P,\alpha)=-\displaystyle\sum_i \displaystyle\sum_{\theta_i} \displaystyle\sum_{x_i}P_{\theta_i}(x_i)\ln(P_{\theta_i}(x_i)) +\displaystyle\sum_i\displaystyle\sum_{\theta}\alpha_i(\theta)\ln(\alpha_i(\theta))$

Note: independent random variables have additive entropies

#### Lagrangian:

$\mathcal{L}= H(P, \alpha) + \displaystyle\sum_i \displaystyle\sum_{\theta_i}\lambda_{\theta_i}\left[ \displaystyle\sum_{x_i}P_{\theta_i}(x_i)-1\right] + \displaystyle\sum_i\lambda_{\alpha_i}\left[\displaystyle\sum_{\theta}\alpha_i(\theta)-1\right]$


$+\displaystyle\sum_{ij}\lambda_{ij}\left[\displaystyle\sum_{\theta}\displaystyle\sum_xP_{\theta_i}(x_i)P_{\theta_j}(x_j)\alpha_i(\theta)u_i(x|\theta_i) -  Y_i(x_i)Y_j(x_j)\alpha_i(\theta)u_i(x|\theta_i)\right]$

### Some notes on the solution

* See the paper for the actual solution
* Maximum entropy (like the empirical likelihood) problems typically take advantage of duality in optimization.  This is why the $p_i>0\;\;$ constraint is not present in the Lagrangian.

```python
from scipy import optimize
import numpy as np

row_payoff = 3
column_payoff= 2


p = np.array([.25,.25,.25,.25,.33,.33,.34])


def obfunc(x):
    return np.sum(x*np.log(x))

def prop1(x):
    return x[0]+x[1]+x[2] +x[3] -1

def prop2(x):
    return x[4] +x[5] +x[6] -1

def ineq1(x):
    return x

def utilcons1(x):
    p1 = x[0:4]
    p2 = x[4:7]
    return np.sum(p1 *np.array([6,5,1,4])* p2[0]) + np.sum(p1 *np.array([2,2,3,1])*p2[1]) + \
                np.sum(p1*np.array([5,1,3,2])*p2[2]) - row_payoff

def utilcons2(x):
    p1 = x[0:4]
    p2 = x[4:7]
    return np.sum(p2 *np.array([3,4,3])* p1[0]) + np.sum(p2 *np.array([4,6,7])*p1[1]) + \
            np.sum(p2*np.array([1,2,1])*p1[2]) + np.sum(p2*np.array([4,3,5])*p1[3])- column_payoff

results = optimize.fmin_slsqp(obfunc, p, eqcons = [prop1, prop2, utilcons1, utilcons2], f_ieqcons= ineq1, 
    epsilon=10**-10, iter=105)

summary=[]
for res in xrange(7):
    summary.append(Eq(allsims[res], round(results[res], 4)))



```
